```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../target/csv/16a0ce28-f5f5-406d-8433-c0d2cbc21cdf.csv")
df = df.drop(columns=['uuid'])
df.groupby(['product'])[['position']].describe()
```

```python
products = df['product'].unique()

for product in products:
    df_product = df[df['product'] == product]
    fig, ax = plt.subplots(figsize=(10, 5))
    ax.set_title(f"Position of {product}")
    ax.scatter(df_product['timestamp'], df_product['position'])
    ax.hlines(y=df_product['position'].mean(), xmin=0, xmax=df_product['timestamp'].max(), linewidth=1, color='r')
    ax.legend(['position', 'mean position'])
```
