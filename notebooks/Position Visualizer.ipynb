```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../target/csv/56bff040-23e8-46f1-9867-20522137707c.log")
df = df.drop(columns=['uuid'])
df.groupby(['product'])[['position']].describe()
```

```python
products = df['product'].unique()

for product in products:
    df_product = df[df['product'] == product]
    fig, ax = plt.subplots(figsize=(10, 5))
    ax.set_title(f"Position of {product}")
    ax.plot(df_product['timestamp'], df_product['position'])
```
